# Import Dependencies

```python
# Import dependencies
import pandas as pd
import numpy as np
```

# Create Dataframes for Independent Variables

```python
# Create a DataFrame for global DDP data. 
globalgdp_df = pd.read_csv("gdp/gdp.csv")
globalgdp_df.head()
```

```python
# Create a DataFrame for global population data. 
globalpopulation_df = pd.read_csv("population/population.csv")
globalpopulation_df.head()
```

```python
# Create a DataFrame for global urbanization data. 
globalurbanization_df = pd.read_csv("urbanization/urbanization.csv")
globalurbanization_df.head()
```

```python
# Extract china gdp from global gdp dataframe. 
chinagdp_df = globalgdp_df.loc[globalgdp_df['Country Name']=="China"]
chinagdp_df.head()
```

```python
# Extract china population from global population dataframe. 
chinapopulation_df = globalpopulation_df.loc[globalgdp_df['Country Name']=="China"]
chinapopulation_df.head()
```

```python
# Extract china urbanization from global urbanization dataframe. 
chinaurbanization_df = globalurbanization_df.loc[globalgdp_df['Country Name']=="China"]
chinaurbanization_df.head()
```

```python
# Concatenate china dataframes into single dataframe
chinadata_df = chinagdp_df.append(chinapopulation_df)
chinadata_df = chinadata_df.append(chinaurbanization_df)
chinadata_df.head()
```

# Create Dataframes for Dependent Variables

```python
# Create a DataFrame for global DDP data. 
globalelectricity_df = pd.read_csv("electricity-generation.csv")
globalelectricity_df.head()
```

```python
# Extract china gdp from global gdp dataframe. 
chinaelectricity_df = globalelectricity_df.loc[globalelectricity_df['Country Name']=="China"]
chinaelectricity_df.head()
```

# Merge Dependent and Independent Variables

```python
# Concatenate china dataframes into single dataframe
alldata_df = chinadata_df.append(chinaelectricity_df)
alldata_df.head()
```

```python
# Select only columns where we have all data
# df.drop(df.loc[:, 'B':'D'].columns, axis = 1)
alldata19852020_df = alldata_df.drop(alldata_df.loc[:, 'Indicator Code':'1984'].columns, axis =1)
alldatadropcolumns_df = alldata19852020_df.drop(alldata19852020_df.loc[:, 'Country Name':'Country Code'].columns, axis =1)
alldataclean_df = alldatadropcolumns_df.drop(alldatadropcolumns_df.loc[:, '2020':'Grand Total'].columns, axis =1)
alldataclean_df.head() 
```

```python
# Transpose clean dataframs for machine learning model
alldataclean_transpose_df = alldataclean_df.transpose()
alldataclean_transpose_df.head()
```

```python
# df.rename(columns={'A':'','B':''})
alldataclean_transpose_df.rename(columns={'Indicator Name':'Year'})
alldataclean_transpose_df.head()
```
